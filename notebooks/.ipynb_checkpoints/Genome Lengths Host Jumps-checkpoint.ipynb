```python
import pandas as pd
import seaborn as sns
```

Get the data out of the big JSON dump file and into a Pandas Dataframe

```python
import re
import json

def parse_file_string(file_string):
    '''gets the elements of the path in the file string of the JSON object'''
    return file_string.split('_')

def get_run_tag(file_string):
    '''gets a run tag that identifies the folder the original pickle was in and allows grouping by run'''
    els = parse_file_string(file_string)
    return "_".join(els[1:-1])
    
def parse_run_info(hash_key):
    '''returns (jump num, generation num, strain num, ) from the hash keys of the JSON object'''
    m = re.match("^jmp(\d+)gen(\d+)strn(\d+)", hash_key)
    return [float(m.group(1)), float(m.group(2)), float(m.group(3))]

def parse_run_results(array):
    '''returns (Num TEs, TE nucleotide length, Num effectors, Effector nucleotide length, Genome length)''' 
    return [array[1], array[0], array[3],array[2],array[4]]

def make_dataframe(JSON_file):
    '''returns a pandas dataframe of the JSON file'''
    jsn = json.loads(open(JSON_file).read())
    files = jsn.keys()

    all_rows = []
    fid = 0
    for f in files:
        fid += 1
        run_tag = get_run_tag(f)
        run_info = [parse_run_info(run) for run in jsn[f].keys()]
        run_results = [parse_run_results(jsn[f][run]) for run in jsn[f].keys()]
        for i in range(len(run_info)):
            all_rows.append( [f, fid, run_tag] + run_info[i] + run_results[i]  )
        
    
    
    return pd.DataFrame(all_rows, columns=('file', 'file_id', 'run_tag', 'jump', 'generation', 'strain','number_TEs','TE_nucleotide_length', \
                               'number_effectors', 'effector_nucleotide_length','genome_length'))
    #results = [parse_run_results(k) for jsnk in jsn.]
        

df = make_dataframe('../cleaned/genome_lengths.json')
df.to_csv("/Users/macleand/Desktop/df.csv")
```
